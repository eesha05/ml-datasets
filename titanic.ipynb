```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
flights=sns.load_dataset('flights')
flights.head()
```

```python
iris=sns.load_dataset('iris')
iris.sample(10)
```

# scatterplot is comparing 2 columns both numerical

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'])
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=df['Sex'])
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

# barplot is used when one is numerical and other is categorical

```python
sns.barplot(x=df['Pclass'],y=df['Age'])
```

```python
sns.barplot(x=df['Pclass'],y=df['Fare'])
```

```python
sns.barplot(x=df['Pclass'],y=df['Age'],hue=df['Sex'])
```

# boxplot is for numerical and categorical

# distplot is for numerical and categorical

```python
sns.distplot(df[df['Survived']==0]['Age'],hist=False)
sns.distplot(df[df['Survived']==1]['Age'],hist=False)
```

# heatmap is used for both categorical

```python
pd.crosstab(df['Pclass'],df['Survived'])
```

```python
sns.heatmap(pd.crosstab(df['Pclass'],df['Survived']))
```

```python
df.groupby('Pclass')['Survived'].mean()*100
```

this is percentage of survival of characters in each class

```python
df.groupby('Sex')['Survived'].mean()*100
```

```python
sns.heatmap(flights.pivot_table(values='passengers',index='month',columns='year'))
```

```python
sns.clustermap(pd.crosstab(df['SibSp'],df['Survived']))
```

```python
sns.clustermap(flights.pivot_table(values='passengers',index='month',columns='year'))
```
